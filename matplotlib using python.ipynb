```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,11)
y=x*2
```

```python
print(x)
y
```

```python
y1=3*x
```

```python
plt.plot(x,y,color="purple",linestyle=":",linewidth=2)
plt.plot(x,y1,color="red",linewidth=2)
plt.title("line plot",color="yellow")
plt.xlabel("pavan")
plt.ylabel("singh")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,color="purple",linestyle=":",linewidth=2)
plt.subplot(2,1,2)
plt.plot(x,y1,color="Green",linestyle=":",linewidth=2)
```

```python
student={"bob":87,"mat":86,"mill":34}
```

```python
name=list(student.keys())
marks=list(student.values())
print(name)
marks
```

```python
plt.bar(name,marks,color="green")
plt.title("std marks")
plt.xlabel("name")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python

plt.barh(name,marks,color="green")
plt.title("std marks")
plt.xlabel("name")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```
